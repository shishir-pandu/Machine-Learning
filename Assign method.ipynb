### 1.1 Feature enginnering:


```python
df=data.copy()
```

```python
def new_columns(df, cols):
    for col in cols:
        df = df.assign(**{
            f'{col}_year': lambda x: x[col].dt.year,
            f'{col}_month': lambda x: x[col].dt.month,
            f'{col}_day': lambda x: x[col].dt.day,
        })
    return df
```

```python
cols=['order_date','ship_date']
new_df=new_columns(df,cols).drop(cols, axis=1)
```

```python
df=new_df.copy()
import matplotlib.pyplot as plt

# create a scatter plot of sales vs profit
plt.scatter(df['sales'], df['profit'])

# set plot title and axis labels
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

# show plot
plt.show()
```

```python
# group the data by category and sum the sales
sales_by_category = df.groupby('category')['sales'].sum()

# create a bar chart of sales by category
sales_by_category.plot(kind='bar')

# set plot title and axis labels
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

# show plot
plt.show()
```

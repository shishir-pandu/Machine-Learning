```python
import pandas as pd
path="melb_data.csv"
```

```python
pd.read_csv(path)
```

```python
data=pd.read_csv(path)
y=data.Price
melb_data=data.drop(['Price'], axis=1)
X=melb_data.select_dtypes(exclude=['object'])
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,train_size=0.7, test_size=0.3)
```

```python
M_XT=X_train.copy()
M_XV=X_valid.copy()
M_yt=y_train.copy()
M_yv=y_valid.copy()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def model(X_train,X_valid,y_train,y_valid):
    model=RandomForestRegressor()
    model.fit(X_train,y_train)
    predections=model.predict(X_valid)
    return mean_absolute_error(predections,y_valid)
```

### 1.Dropping the missing values

```python
cols_missing_data=[
    col for col in M_XT
    if M_XT[col].isnull().any()
]
cols_missing_data
```

```python
M_XT_Dropped=M_XT.drop(cols_missing_data,axis=1)
M_XV_Dropped=M_XV.drop(cols_missing_data,axis=1)
```

```python
print("MAE calculated from the dropping table is:", model(M_XT_Dropped,M_XV_Dropped,y_train,y_valid))
```

### 2.By imputing the missing values:

```python
from sklearn.impute import SimpleImputer
```

```python
myimputer=SimpleImputer()
M_XT_I=pd.DataFrame(myimputer.fit_transform(M_XT))
M_XV_I=pd.DataFrame(myimputer.fit_transform(M_XV))
```

```python
print("MAE calculated from the imputation is:", model(M_XT_I,M_XV_I,y_train,y_valid))
```

### 3.Extension of imputation

```python
myImputer=SimpleImputer()
for col in cols_missing_data:
    M_XT[col +"_was_missing"]=M_XT[col].isnull()
    M_XV[col +"_was_missing"]=M_XV[col].isnull()
```

```python
M_XT_EI=pd.DataFrame(myImputer.fit_transform(M_XT))
M_XV_EI=pd.DataFrame(myImputer.transform(M_XV))
```

```python
print("MAE calculated from the Extension of Imputation is:", model(M_XT_I,M_XV_I,y_train,y_valid))
```
